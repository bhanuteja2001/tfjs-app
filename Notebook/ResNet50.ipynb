```python
data = []
image_size = 224
test_pixel_data = []
import matplotlib.pyplot as plt
from PIL import Image
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True


for i in range(len(Train)):
    img_array = Image.open(Train['file'][i])
    new_img_array = np.array(img_array)
    if new_img_array.ndim != 3:
        new_img_array = cv2.cvtColor(new_img_array, cv2.COLOR_GRAY2BGR)
    new_img_array = cv2.resize(new_img_array,(224,224))
    if new_img_array.shape[2] != 3:
        new_img_array = new_img_array[:, :, :3]
    print(i,np.array(new_img_array).shape,Train['Label'][i])
    data.append([new_img_array, Train['Label'][i]])
```

```python
for i in range(len(Test)):
    img_array = Image.open(Test['file'][i])
    new_img_array = np.array(img_array)
    if new_img_array.ndim != 3:
        new_img_array = cv2.cvtColor(new_img_array, cv2.COLOR_GRAY2BGR)
    new_img_array = cv2.resize(new_img_array,(224,224))
    if new_img_array.shape[2] != 3:
        new_img_array = new_img_array[:, :, :3]
    print(i,np.array(new_img_array).shape,Test['Label'][i])
    test_pixel_data.append(new_img_array)
```

```python
np.random.shuffle(data)
x = []
y = []

counter = 406
count = 0

for image in data:
    try:
        x.append(image[0])
        y.append(image[1])
    except:
        print(count)
    count += 1
```

```python
# converting x & y to numpy array as they are list
x = np.array(x)
y = np.array(y)
```

```python
pd.Series(y).value_counts()
```

```python
X_train, X_val, y_train, y_val = train_test_split(x,y,test_size=0.2, random_state = 42,stratify=y)
```

```python
from keras.utils.np_utils import to_categorical
y_train = to_categorical(y_train)
y_val = to_categorical(y_val)
```

```python
test_pixel_data = np.array(test_pixel_data)

test_pixel_data = test_pixel_data/255

print(X_train.shape,X_val.shape,test_pixel_data.shape)
```

```python
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1.0/255.0)

val_datagen = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1.0/255.0)

train_generator = train_datagen.flow(X_train,y_train,batch_size=8)

val_generator = val_datagen.flow(X_val,y_val,batch_size=8)
```

```python
print(X_train.shape,y_train.shape,X_val.shape)
```

```python
from tensorflow.keras.applications import ResNet50
from keras.layers import *
from keras.models import Sequential
from keras.regularizers import *
from tensorflow.keras.optimizers import Adam
```

```python
def build_model():
    model = Sequential()

    conv_base = ResNet50(input_shape=(224,224,3), include_top=False, pooling='max',weights='imagenet')
    model.add(conv_base)
    model.add(BatchNormalization())
    model.add(Dense(2048, activation='relu', kernel_regularizer=l1_l2(0.01)))
    model.add(BatchNormalization())
  # model.add(Dense(2048, activation='relu', kernel_regularizer=l1_l2(0.01)))
  # model.add(BatchNormalization())
  # model.add(Dense(2048, activation='relu', kernel_regularizer=l1_l2(0.01)))
  # model.add(BatchNormalization())
  # model.add(Dense(1024, activation='relu', kernel_regularizer=l1_l2(0.01)))
  # model.add(BatchNormalization())
    model.add(Dense(5, activation='softmax'))
    for layer in model.layers:
        layer.trainable = True
 
    model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=1e-4), metrics=['accuracy',"Precision","Recall"])

    return model
```

```python
cbs = [tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, min_lr=1e-5, verbose=0),
tf.keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=1e-6, patience=5, verbose=1, mode='auto')]
```

```python
model = build_model()
model.summary()
```

```python
from time import time
```

```python
%%timeit -n1 -r1 
with tf.device("GPU"):
    model.fit(train_generator,steps_per_epoch=87,epochs=100,validation_data=val_generator,validation_steps=22,callbacks=[cbs])
```

```python
fig, ax  = plt.subplots(2,1)

#Loss
ax[0].plot(model.history.history['loss'], color = 'b', label='Training Loss')
ax[0].plot(model.history.history['val_loss'], color = 'r', label='Validation Loss')
ax[0].legend()
#Accuracy
ax[1].plot(model.history.history['accuracy'], color='b', label='Training Accuraccy')
ax[1].plot(model.history.history['val_accuracy'], color='r', label='Val Accuracy')
ax[1].legend()
```
